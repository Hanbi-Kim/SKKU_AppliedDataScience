# K 값을 선택하는 방법
- K 값이 작을수록 Variance가 크고 (Less stable)/ K 값이 클수록 Bias가 커진다.(Less Precise)
- Cross-Validation을 통해 검증에러를 최소화 할수있게 선택

# KNN의 장점
- 사전에 다른 훈련이 필요없시 주어진 데이터만으로 수행가능
- 모든 데이터를 훈련에 이용하여 정보 손실을 최소화 할수있음
- 알고리즘을 이해하기가 쉬움

# KNN의 단점
- 이상치에 매우 민감하게 반응
- 훈련 데이터가 불균형 데이터인 경우, Majority 다수의 개체를 가진 클래스에 집중적으로 반응
- 모델을 저장하지 않고 모든 데이터를 활용함으로 저장 공간이 많이 필요
- 훈련 데이터가 바뀔시 결과가 전체적으로 바뀜
- 새로운 데이터가 들어갈시 모든 거리를 다시 계산해야 함으로 많은 시간이 소요
- 차원이 높아질수록 거리가 가깝다는 의미를 잃게됨 (Curse of Dimensionality) 

# KNN의 단점을 보완하는 방법
- PCA 등을 통해 차원이 높은 경우 차원을 줄이고 KNN을 진행
- Sampling/Clustering을 통해 데이터가 많은 경우 데이터의 양을 줄이고 진행 
- Space-Partitioning을 통해 공간을 나누어서 진행 

# Overview

KNN 알고리즘은 이해하기 쉬워 처음에 적용하기 쉽지만, 전체적인 활용도면에서는 단점이 많은 알고리즘이라고 할 수 있다. 
이번 데이터 Iris의 경우, 시각화에서도 보이듯, 세 가지 종이 전체적으로 구분이 잘되는 것을 볼 수 있다. 따라서 해당 데이터를 분류하는 데에는 문제가 없어 보이지만, 데이터가 좀 더 복잡해지고 차원이 높아질 경우 비슷한 성능을 낼수 있다고 하기는 힘들다.
